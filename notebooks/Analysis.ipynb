# LLM Crystallography Evaluation — Analysis

This notebook analyses the performance of Claude on structured crystallography reasoning tasks.

The evaluation is based on:
- rubric-based scoring
- multiple prompt types
- domain-specific reasoning criteria

```python
import pandas as pd
```

```python
results = pd.read_csv("../data/results_manual.csv")
scores = pd.read_csv("../data/scores.csv")

# Clean columns
results.columns = results.columns.str.strip()
scores.columns = scores.columns.str.strip()
```

```python
results["response_length"] = results["response_text"].apply(
    lambda x: len(str(x).split())
)
```

```python
merged = results.merge(
    scores,
    on=["model", "module", "prompt_type", "prompt_id"],
    how="inner"
)
```

```python
merged["total_score"].describe()
```

### Overall Performance

Summary statistics of total scores across all prompts.

```python
merged["total_score"].describe()
```

```python
merged.groupby("prompt_type")["total_score"].mean()
```

### Performance by Prompt Type

Strategy and prioritisation prompts perform better than open-ended ones.

```python
merged.groupby("module")["total_score"].mean()
```

### Performance by Task Type

Peak position tasks show weaker performance compared to other modules.

```python
merged.groupby("prompt_type")["response_length"].mean()
```

```python
merged[["response_length", "total_score"]].corr()
```

### Verbosity vs Performance

Examines whether longer responses correlate with higher scores.

## Key Insights

- Performance is highest on structured reasoning tasks (prioritisation, critique)
- Open-ended prompts lead to weaker, less focused answers
- Certain domains (e.g. peak positions) are more challenging for the model
- Response length shows limited correlation with score — verbosity ≠ quality

These results highlight strengths in structured reasoning and weaknesses in precise physical interpretation.
